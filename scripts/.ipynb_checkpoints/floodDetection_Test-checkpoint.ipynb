```python
import ee
import geemap
import matplotlib.pyplot as plt
import numpy as np
```

```python
Map = geemap.Map()
```

```python
time_start = '2022-01-01'
time_end = '2023-01-01'
```

```python
# test area
geometry = ee.Geometry.Polygon([
    [[68.06523955643267,27.402689139647947],
     [68.68802703201861,27.402689139647947],
     [68.68802703201861,27.925089418721566],
     [68.06523955643267,27.925089418721566],
     [68.06523955643267,27.402689139647947]]
])
```

```python
Map.centerObject(geometry, 8)
```

```python
# function to reduce speckle noise
def despeckel(img):
    return img.focal_mean(30, 'square', 'meters').copyProperties(img, img.propertyNames())
```

```python
image_collection = ee.ImageCollection("COPERNICUS/S1_GRD")
```

```python
before = (image_collection
    .filterDate(time_start, time_end)
    .filter(ee.Filter.calendarRange(3, 3, 'month'))
    .filterBounds(geometry)
    .filter(ee.Filter.listContains('transmitterReceiverPolarisation', 'VV'))
    .filter(ee.Filter.eq('instrumentMode', 'IW'))
    .select('VV')
    .map(despeckel)
    .min())

# Add the 'before' layer to the map
Map.addLayer(before.clip(geometry), {}, 'Before', False)
```

```python
after = (image_collection
    .filterDate(time_start, time_end)
    .filter(ee.Filter.calendarRange(8, 8, 'month'))
    .filterBounds(geometry)
    .filter(ee.Filter.listContains('transmitterReceiverPolarisation', 'VV'))
    .filter(ee.Filter.eq('instrumentMode', 'IW'))
    .select('VV')
    .map(despeckel)
    .min())

# Add the 'after' layer to the map
Map.addLayer(after.clip(geometry), {}, 'After', False)
```

```python
# calculate change detection
change = before.subtract(after)

# adiding the change layer to the map with 90% stretch
change_min = change.reduceRegion(
    reducer=ee.Reducer.percentile([10]),
    geometry=geometry,
    scale=30
).get('VV')

change_max = change.reduceRegion(
    reducer=ee.Reducer.percentile([90]),
    geometry=geometry,
    scale=30
).get('VV')

Map.addLayer(change.clip(geometry), 
              {'min': change_min, 'max': change_max}, 
              'Flooded Region', False)
```

```python
# change values for the histogram
change_data = change.reduceRegion(
    reducer=ee.Reducer.toList(),
    geometry=geometry,
    scale=30,
    maxPixels=1e9
).getInfo()

# extract the list of values
change_values = change_data['VV']

# plot the histogram using matplotlib
plt.figure(figsize=(10, 6))
plt.hist(change_values, bins=30, color='blue', edgecolor='black')
plt.title('Histogram of Change Detection')
plt.xlabel('Change Value')
plt.ylabel('Frequency')
plt.show()
```

```python
threshold = 6.5
```

```python
threshold_before = -20
threshold_after = -20
```

```python
# add thresholded flood layers to the map
Map.addLayer(change.gt(threshold).clip(geometry), {}, 'Flood Threshold', False)
#before = before.gt(-16).clip(geometry)
Map.addLayer(before.gt(threshold_before).clip(geometry), {}, 'Water Before', False)
#after = after.gt(-16).clip(geometry)
Map.addLayer(after.gt(threshold_after).clip(geometry), {}, 'Water After', False)
```

```python
imageCollection = ee.ImageCollection('COPERNICUS/S1_GRD')  # Example image collection

before = imageCollection.filterDate(time_start, time_end) \
    .filterBounds(geometry) \
    .filter(ee.Filter.calendarRange(3, 3, 'month')) \
    .filter(ee.Filter.listContains('transmitterReceiverPolarisation', 'VV')) \
    .filter(ee.Filter.eq('instrumentMode', 'IW')) \
    .select('VV') \
    .min() \
    .lt(threshold_before)

after = imageCollection.filterDate(time_start, time_end) \
    .filterBounds(geometry) \
    .filter(ee.Filter.calendarRange(8, 8, 'month')) \
    .filter(ee.Filter.listContains('transmitterReceiverPolarisation', 'VV')) \
    .filter(ee.Filter.eq('instrumentMode', 'IW')) \
    .select('VV') \
    .min() \
    .lt(threshold_after)

# mask the images to show only values equal to 1
before_masked = before.updateMask(before)
after_masked = after.updateMask(after)

# gettinhg the masked images as numpy arrays
before_array = geemap.ee_to_numpy(before_masked, region=geometry.bounds(), scale=30) 
after_array = geemap.ee_to_numpy(after_masked, region=geometry.bounds(), scale=30)

# plotting the images side by side using matplotlib
fig, axes = plt.subplots(1, 2, figsize=(15, 7))

# Plot 'before' image
axes[0].imshow(before_array, cmap='Blues', vmin=0, vmax=1)
axes[0].set_title('Water Before Flooding')
axes[0].axis('off') 

# Plot 'after' image
axes[1].imshow(after_array, cmap='Blues', vmin=0, vmax=1)
axes[1].set_title('Water After Flooding')
axes[1].axis('off')

plt.tight_layout()
plt.show()
```
